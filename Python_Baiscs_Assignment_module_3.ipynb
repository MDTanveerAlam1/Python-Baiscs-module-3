```python
# Question Number 2 >>>
# Describe the role of predefined keywords in Python and provide examples of how they are used in a program ?
# Answer >>

# Python keywords are reserved words that have special meanings and purposes within the language.
# They cannot be used as variable names or identifiers.  They form the building blocks of Python syntax.

# Examples >>>

# 1. if, elif, else: Control the flow of execution based on conditions >>
x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")

# 2. for, while:  Used for creating loops to repeat blocks of code >>
for i in range(5): # range is also a keyword-related function
    print(i)
i = 0
while i < 5:
    print(i)
    i += 1

# 3. def: Defines a function >>
def my_function():
    print("Hello from a function")

my_function()


# 4. return: Exits a function and optionally returns a value >>
def add(a, b):
    return a + b

result = add(3, 4)
print(result)

# 5. import: Used to bring in external modules or libraries >>
import math # math is a module, import is a keyword
print(math.sqrt(25))

# 6. class: Used to define a class, which is a blueprint for creating objects >>
class Dog:  # Class definition
    def __init__(self, name):
        self.name = name

    def bark(self):
        print("Woof!")

my_dog = Dog("Buddy") # Object creation
my_dog.bark() # Object method call

# 7. try, except: Used for exception handling (managing errors) >>
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: Division by zero")


# These are just a few examples.  Other important keywords include:
# and, or, not, True, False, None, break, continue, pass, in, is, del, global, nonlocal... and many more.
# Each serves a specific function within the Python language.
```

```python
# Question Number 4 >>>
# Discuss the different types of operators in Python and provide examples of how they are used ?
# Answer >>


# Arithmetic Operators ->> Perform mathematical calculations >>
# + (Addition)
# - (Subtraction)
# * (Multiplication)
# / (Division)
# % (Modulo - remainder)
# // (Floor Division - quotient)
# ** (Exponentiation)

x = 10
y = 3
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x % y)
print(x // y)
print(x ** y)


# Comparison Operators ->> Compare two values.
# == (Equal to)
# != (Not equal to)
# > (Greater than)
# < (Less than)
# >= (Greater than or equal to)
# <= (Less than or equal to)

a = 5
b = 10
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)


# Comparison Operators ->> Compare two values.
# == (Equal to)
# != (Not equal to)
# > (Greater than)
# < (Less than)
# >= (Greater than or equal to)
# <= (Less than or equal to)

a = 5
b = 10
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)


# Logical Operators ->> Combine conditional statements.
# and (True if both operands are true)
# or (True if at least one operand is true)
# not (Inverts the truth value)

p = True
q = False
print(p and q)
print(p or q)
print(not p)


# Assignment Operators ->> Assign values to variables.
# = (Assignment)
# += (Add and assign)
# -= (Subtract and assign)
# *= (Multiply and assign)
# /= (Divide and assign)
# %= (Modulo and assign)
# //= (Floor divide and assign)
# **= (Exponentiate and assign)

num = 5
num += 2
print(num)


# Bitwise Operators ->> Perform operations on individual bits of binary numbers.
# & (Bitwise AND)
# | (Bitwise OR)
# ^ (Bitwise XOR)
# ~ (Bitwise NOT)
# << (Left shift)
# >> (Right shift)

# Example (Bitwise AND)
n1 = 10  # Binary: 1010
n2 = 4   # Binary: 0100
result = n1 & n2 # Binary: 0000
print(result)


# Membership Operators ->> Test if a value is present in a sequence (string, list, tuple, set, dictionary).
# in (True if value is found)
# not in (True if value is not found)

my_list = [1, 2, 3, 4, 5]
print(3 in my_list)  # Output: True
print(6 not in my_list) # Output: True


# Identity Operators ->> Compare the memory locations of two objects.
# is (True if both variables refer to the same object)
# is not (True if both variables refer to different objects)

x = 5
y = 5
print(x is y)

a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)
```

```python
# Question Number 5 >>>
#  Explain the concept of type casting in Python with examples ?
# Answer >>

# Type casting in Python is the process of converting a value from one data type to another.  It's like changing the "shape" of the data.

# Implicit Type Casting (Automatic) ->> Python sometimes does this for you.
# Example >>
x = 10  # Integer
y = 3.14  # Float
z = x + y  # Python automatically converts x to a float before adding

print(z) # The result is a float because of implicit type conversion.


# Explicit Type Casting (Manual) ->> You tell Python to convert the data type using built-in functions.

# 1. int() ->> Converts a value to an integer.  If a float is given, the decimal part is discarded.

num_str = "25"
num_int = int(num_str)
print(num_int) # (an integer)

num_float = 3.8
num_int_from_float = int(num_float)
print(num_int_from_float)  # (decimal part is truncated)

# 2. float() ->>  Converts a value to a floating-point number.

num_int_2 = 50
num_float_2 = float(num_int_2)
print(num_float_2)


# 3. str() ->> Converts a value to a string.

num_int_3 = 75
num_str_2 = str(num_int_3)
print(num_str_2 + " is my lucky number") # (concatenation as strings)

# 4. bool() ->> Converts a value to a Boolean (True or False).

# Values that are considered "falsy" -> 0, empty strings, empty lists, None, etc.
# Other values are considered "truthy."
val = 0
print(bool(val))

val2 = 10
print(bool(val2))

my_list = [] # An empty list is falsy
print(bool(my_list))

my_list.append(1) # Not empty any more
print(bool(my_list))

# Why is type casting useful?

# 1. Input Handling ->> User input is often treated as strings, but we may need to convert them to numbers for calculations.

# 2. Data Combination ->>  Combine data from different sources (like string and number).

# 3. Function Requirements ->> Functions might need specific data types, such as converting an integer into string if the function only accepts string parameters.
```

```python
# Question Number 6 >>>
# How do conditional statements work in Python ? Illustrate with examples ?
# Answer >>

# Conditional statements in Python control the flow of your program based on whether certain conditions are true or false.

# 1. if statement ->> Executes a block of code only if a condition is true.
age = 20
if age >= 18:
    print("You are an adult.")

# 2. if-else statement ->> Executes one block of code if a condition is true and another block if it's false.
temperature = 25
if temperature > 30:
    print("It's a hot day!")
else:
    print("It's not too hot.")

# 3. if-elif-else statement ->> Allows for multiple conditions to be checked sequentially. The first condition that evaluates to true triggers its corresponding code block.
grade = 85
if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")
else:
    print("F")

# 4. Nested Conditional Statements ->>  You can place conditional statements inside other conditional statements.
num = 15
if num > 10:
  if num % 2 == 0:
     print("Number is greater than 10 and even")
  else:
     print("Number is greater than 10 and odd")
else:
   print("Number is not greater than 10")
```

```python
# Question Number 7 >>>
# Describe the different types of loops in Python and their use cases with examples.
# Answer >>

# 1. for loop ->> Used for iterating over a sequence (like a list, tuple, string, or range) or other iterable objects.

# Example >>> Iterating through a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Example >>> Iterating through a string
message = "Hello"
for letter in message:
    print(letter)

# Example >>> Iterating through a range of numbers
for i in range(5):  # Prints numbers 0 to 4
    print(i)


# 2. while loop ->>  Repeats a block of code as long as a condition is true.  You must make sure the condition eventually becomes false to avoid an infinite loop!

# Example >>> Counting to 5
count = 0
while count < 5:
    print(count)
    count += 1  # Increment the counter, or you'll have an infinite loop

# Example >>> Reading input until a specific value is entered:
user_input = ""
while user_input != "quit":
    user_input = input("Enter something ('quit' to exit): ")
    print("You entered:", user_input)


# 3. Nested Loops ->> Loops within loops. The inner loop runs completely for each iteration of the outer loop.

# Example >>> Creating a multiplication table
for i in range(1, 6):  # Outer loop: Rows
    for j in range(1, 6):  # Inner loop: Columns
        print(i * j, end="\t")  # Print the product, use \t for tab spacing
    print() # move to next line after each row is printed

# <<- Loop Control Statements ->>

# break ->> Exits the loop immediately.
# Example >>>
for i in range(10):
  if i == 5:
    break # stop at 5
  print(i)

# continue ->> Skips the rest of the current iteration and moves to the next one.
# Example >>>
for i in range(10):
  if i % 2 == 0:
    continue # skip all even numbers
  print(i)

# pass ->>  Does nothing.  Often used as a placeholder when you need syntactically correct code but don't want any action to take place.
# Example >>>
for i in range(5):
    pass # Does nothing; can be used when you need a placeholder for future code
```
